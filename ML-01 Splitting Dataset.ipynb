```python
import os
import shutil
import random

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dir = '/content/drive/MyDrive/Capstone Project/ML Colab/Dataset/Tanah'
train = '/content/drive/MyDrive/Capstone Project/ML Colab/Dataset/Fix/train/'
test = '/content/drive/MyDrive/Capstone Project/ML Colab/Dataset/Fix/test/'
val = '/content/drive/MyDrive/Capstone Project/ML Colab/Dataset/Fix/val/'

train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

for label in os.listdir(dir):
  label_dir = os.path.join(dir, label)
  img = os.listdir(label_dir)
  random.shuffle(img)
  img_len = len(img)
  train_split = int(img_len * train_ratio)
  val_split = int(img_len * (train_ratio + val_ratio))

  for i in img[:train_split]:
    src = os.path.join(label_dir, i)
    dst = os.path.join(train, label, i)
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy(src, dst)

  for i in img[train_split:val_split]:
    src = os.path.join(label_dir, i)
    dst = os.path.join(val, label, i)
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy(src, dst)

  for i in img[val_split:]:
    src = os.path.join(label_dir, i)
    dst = os.path.join(test, label, i)
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy(src, dst)
```
